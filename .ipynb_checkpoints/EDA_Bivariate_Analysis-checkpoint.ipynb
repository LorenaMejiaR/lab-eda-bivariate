```python
contingency_table = pd.crosstab(df['category'], df['isBestSeller'])
```

```python
# 2. Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
```

```python
print(chi2_stat)
print(p_value)
print(dof)
print(expected)
```

```python
# 3. Compute Cramér's V to understand the strength of association between best-seller status and category.
n = contingency_table.sum().sum()  
min_dim = min(contingency_table.shape)
cramers_v = np.sqrt(chi2_stat / (n * (min_dim - 1)))

print(cramers_v)
```

```python
# 4. Visualize the relationship between product categories and the best-seller status using a stacked bar chart.
ax = contingency_table.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(8, 6))

ax.set_xlabel('Product Category')
ax.set_ylabel('Proportion')
ax.set_title('Proportion of Best-Sellers by Product Category')
ax.legend(title='Is Best-Seller', labels=['No', 'Yes'])

plt.show()
```

# Part 2: Exploring Product Prices and Ratings Across Categories and Brands

```python
# Remove the outliers 
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
df_cleaned
```

```python
# 1. Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.
category_counts = df['category'].value_counts()
top_categories = category_counts.head(20).index
top_categories
```

```python
df_top_categories = df[df['category'].isin(top_categories)]
df_top_categories
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='category', y='price', data=df_top_categories, palette='Set2')

plt.title('Distribution of Product Prices across Top 20 Categories', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Product Price', fontsize=12)

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
#2. Which product category tends to have the highest median price? Don't filter here by top categories.
category_medians = df.groupby('category')['price'].median()
sorted_category_medians = category_medians.sort_values(ascending=False)
sorted_category_medians
```

## After checking the median prices of the categories, we can see that Laptops tends to have the highest median price. 

```python
# 3. Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
category_counts = df['category'].value_counts()
top_categories = category_counts.head(10).index  # Get the top 10 categories
top_categories
```

```python
df_top_categories = df[df['category'].isin(top_categories)]
category_avg_prices = df_top_categories.groupby('category')['price'].mean()
category_avg_prices
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=category_avg_prices.index, y=category_avg_prices.values, palette='Set2')

# Set the title and labels
plt.title('Average Product Price by Top 10 Categories', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Average Price', fontsize=12)

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 4. Which product category commands the highest average price? Don't filter here by top categories.
category_means = df.groupby('category')['price'].mean()
sorted_category_means = category_means.sort_values(ascending=False)
sorted_category_means
```

## After checking the mean prices of the categories, we can see that Laptops tends to have the highest mean price. 

```python
# 5. Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.

plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=df_top_categories, palette='Set2')

# Set the title and labels
plt.title('Distribution of Product Ratings by Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Product Rating', fontsize=12)

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Which category tends to receive the highest median rating from customers? Don't filter here by top categoriess = df.groupby('category')['price'].mean()
category_median_stars = df.groupby('category')['stars'].median()
sorted_category_median_stars = category_median_stars.sort_values(ascending=False)
sorted_category_median_stars
```

# Part 3: Investigating the Interplay Between Product Prices and Ratings

```python
# 1. Calculate the correlation coefficient between price and stars.
correlation = df['price'].corr(df['stars'])
correlation
```

## The Pearson correlation coefficient of -0.1249 suggests a weak negative correlation between product price and product rating.

```python
# 2. Use a scatter plot to visualize the relationship between product rating and price. 

plt.figure(figsize=(10, 6))
sns.scatterplot(x='stars', y='price', data=df, color='blue', edgecolor='black', s=100)

plt.title('Scatter Plot of Product Rating vs Price', fontsize=16)
plt.xlabel('Product Price', fontsize=12)
plt.ylabel('Product Rating', fontsize=12)

plt.tight_layout()
plt.show()
```

## As we can see in the scatter plot, there is not a pattern between the product rating (stars) and the prices.

```python
# 4. Examine if product prices typically follow a normal distribution using a QQ plot.
plt.figure(figsize=(8, 6))
stats.probplot(df['price'], dist="norm", plot=plt)

plt.title('QQ Plot of Product Prices', fontsize=16)
plt.tight_layout()

plt.show()
```
